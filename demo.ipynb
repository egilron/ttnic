## Python med innebygde og eksterne moduler

```python
## Vi kan gjøre en hel del med kjerne-Python uten tillegg
print("Hello, World")
print()
liste_med_tall = (10,11,22,18,13,19, 14, 15)
print("Vi skal jobbe med denne lista:", liste_med_tall)
print("Summen av alle tallene i lista er", sum(liste_med_tall))
print("Det høyeste tallet i lista er", max(liste_med_tall))
print()
#Iterasjon: VI går gjennom ei liste / rekke med noe og gjør noe med hvert element
for tall in liste_med_tall[:3]:  # De tre første elementene i lista
    print("Det dobbelte av", tall, "er", 2 * tall)
```

```python
## Noen moduler kommer alltid sammen med Python, men de må importeres først
import math
print("Roten av 9 er ", math.sqrt(9))

import datetime
now = datetime.datetime.now()
print ("Klokka er", now.strftime("%H:%M den %d-%m-%Y "))

import itertools
noen_tall = [4,5,19]
noen_bokstaver = ["D", "K"]
print("\n Alle kombinasjoner av ett ellement fra hver av listene", noen_bokstaver, "og", noen_tall)
for kombinasjon in itertools.product(noen_tall, noen_bokstaver):
    print(kombinasjon)
```

Andre moduler/ pakker må installeres før de kan importeres. Det er veldig mange av disse, og det er en viktig del at et python-prosjekt å finne ut hvilke moduler man trenger (Det er kanskje et annet korrekt ord for "moduler" på norsk, som jeg ikke kjenner til). 
Anaconda er en tjeneste som pakker et stort antall med slike tilleggsmoduler for Python inn i en samlet nedlasting. Når man har fått inn den pakka, trenger man ikke gå ut på internett for hver ny pakke man skal ha. Utvalget i Anaconda er bredt, mne de har en fokus på data science, forskining og maskinlæring.

```python
## Pakker jeg har installert fra Anaconda på min maskin, og derfor kan importere inn i et Python-program
import numpy as np
print ("Standardavvik i vår liste med tall er", round(np.std(liste_med_tall),2))
print("Det var det numpy som hjalp oss med")
```
